# Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('nba.csv')
df
```

```python
df.isna().sum()/len(df)*100
```

```python
# We drop all the rows that contain null values, we now are not focusing on cleaning te data.
# We want tot explain the concept of groupby()
df.dropna(inplace = True)
df
```

`groupby()` method is used to split a DataFrame into group based on one or more columns, allowing for efficient data analysis.
> It is follows a "split-apply-combine" strategy. That is mean the data is divided into groups,
>  a function is applied to each group, and the result is combined to a new DataFrame.


## Splitting: Dividing the DataFrame into group based on some criteria.

```python
aggregation_data = df.groupby('Team')   # Group the dataset by the 'Team' column
aggregation_data.first()                # "first()" method return the first row for each team,
                                        # it is useful when we have duplicated teams.
```

```python
# We cn get th last value in each team also.
aggregation_data.last()
```

```python
# We can print two or more names in each team.
aggregation_data.apply(lambda x:x.head(2))
```

```python
# We can also apply multiple columns.
aggregation_data = df.groupby(['Team', 'Position'])
aggregation_data.first()
```

## Applying: A function is applied to each group

1- `Aggreration`: calculate summary statistics

```python
aggregation_data = df.groupby(['Team', 'Position']).agg(
    Total_Salary = ('Salary','sum'),
    Avg_Salary = ('Salary','mean'),
    Player_Count = ('Name','count')
)
aggregation_data
```

2- `Transformation`: It modify the values within each group.

```python
""" Purpose: Apply the specific operation while maintaining the original shape of the dataset.
Unlike aggregation which reduces the data.
"""
df['Rank'] = df.groupby('Team')['Salary'].transform(lambda x : x.rank(ascending = False, method = 'max'))
df
```

3- `Filtering`: It keep or discard groups based on certain conditions 

```python
# It allows us to drop entire groups based on a condition. It is useful in cleaning the data by removing specific groups.
df_filtered = df.groupby('Name').filter(lambda x:x['Salary'].mean() >= 1000000)
df_filtered
```
